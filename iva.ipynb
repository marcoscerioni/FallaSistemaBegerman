```
def lee_entero():
    opcion = 0
    print('Presiona 0 para Finalizar, 1 para Ventas, 2 para Compras: ')

    while (True):
        opcion = input()
        try:
            opcion = int(opcion)
        except ValueError:
            print ("Incorrecto, intente nuevamente. ")
        if opcion == 1:
            ventas()
        elif opcion == 2:
            compras()
        elif opcion == 0:
            print('Hasta la próxima')
            break
        else: 
            opcion = 0
            print ("Ingrese un valor valido: ")


def compras():
    print('Vamos a cambiar el archivo de Compras...')
    print('Escriba tal cual el nombre del archivo, con la extensión.')
    print('Por ejemplo: archivo.TXT: ')

    archivo = input()
    with open(archivo, 'r') as file:
        line = file.read()
    cantidad = int(len(line) / 325)
    print('En este archivo hay ' + str(cantidad) + ' compras\n')

    lista = list(line)   # Lo transformo en array para trabajar.
    offset = 0
    for i in range(1, cantidad+1):
        lista[offset + 238] = ' '
        offset = offset + 326

    final = ''.join(lista)
    with open(archivo, 'w') as cambiar:
        cambiar.write(final)
    print('LISTO ARCHIVO RESUELTO :D\n')
    lee_entero()
    return final


def ventas():
    print('Vamos a cambiar el arhivo de Ventas...')
    print('Escriba tal cual el nombre del archivo, con la extensión.')
    print('Por ejemplo: archivo.TXT: ')

    archivo = input()
    with open(archivo, 'r') as file:
        line = file.read()
    cantidad = int(len(line) / 266)
    print('En este archivo hay ' + str(cantidad) + ' compras\n')

    lista = list(line)   # Lo transformo en array para trabajar.
    offset = 0
    for i in range(1, cantidad+1):
        lista[offset + 242] = ' '
        offset = offset + 268

    offset = 266
    for i in range(1, cantidad+1):
        lista[offset] = ''
        offset = offset + 268
    
    final = ''.join(lista)
    with open(archivo, 'w') as cambiar:
        cambiar.write(final)
    print('LISTO ARCHIVO RESUELTO\n')
    lee_entero()

    return final

lee_entero()
```
